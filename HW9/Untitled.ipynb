```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
with open('mini-data.txt', 'r') as file:
    lines = file.readlines()

# Extract features and labels
data = np.loadtxt("mini-data.txt")
X = data[:, :2]  # Features (2D data points)
y = data[:, 2]   # Labels (0 or 1)

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('Scatter Plot of Mini Data Set')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
tree_model = DecisionTreeClassifier(max_depth=3)  # You should replace this with your actual trained model
tree_model.fit(X,y)  # Make sure to replace X_train and y_train with your training data

# Display the decision tree
plt.figure(figsize=(10, 8))
plot_tree(tree_model, filled=True, feature_names=['Feature1', 'Feature2'], class_names=['0', '1'])
plt.show()
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


# Fit AdaBoost with Decision Stumps
n_stumps = 5  # Adjust the number of stumps as needed
ada_model = AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1), n_estimators=n_stumps)
ada_model.fit(X, y)

# Display each stump
for i, tree_in_ensemble in enumerate(ada_model.estimators_):
    plt.figure(figsize=(12, 8))
    plot_tree(tree_in_ensemble, filled=True, feature_names=['Feature1', 'Feature2'], class_names=['0', '1'])
    plt.title(f'Decision Stump {i + 1}')
    plt.show()
```

```python
train_accuracies = [accuracy_score(y, pred) for pred in ada_model.staged_predict(X)]
print("Stump\tAccuracy")
for i, acc in enumerate(train_accuracies):
    print(f"{i+1}\t{acc:.4f}")
```
